<div class="alert alert-secondary" id="What-is-JSON">    
    <span class="topic"> &#10070; What is JSON</span>   
</div>

<br>

<kbd>JSON</kbd> stands for **J**ava**s**cript **O**bject **N**otation. It is a lightweight format to store data that needs to be transported across the internet. The format has been designed in such a way that it is interchangeable with a JavaScript Object. JavaScript is the language of the browser. In case you are wondering why are we talking about another language in a Python class, it is because, the Internet evolved around the browser. And as such, JavaScript became an integral part of the data formats that are used in communication across the internet. Luckily, you don't have to learn JavaScript to understand JSON. Most languages, including Python can handle JSON data. 

How does a sample JSON format look like ?

```python
name = {
        "first_name" : "John", 
        "last_name" : "Doe"
       }
```

Essentially, these are **Key** , **Value** pairs. 

<br>

<div class="text-center">
    
</div>

<br>

This is an example of a JSON format. Well, this is a Python Dictionary, you might say. That is true as well. The Python dictionary is exactly similar to the JSON format. You might have another question here - Why does Python have dictionary as a data type but not JSON ? That is a good question and the answer lies in the fact that Python is not designed as an "Internet First" language like JavaScript. Python is a general purpose programming language that can do many different things. So, Python's data types are designed to solve most general purpose problems. 

However, Python's dictionary data structure happens to be a very good fit for the JSON data format. And that's what Python will be using as the data type to store JSON data. Python has an inbuilt module to work with JSON data.

```python
import json
```

That's all we need to start working with JSON data. Let's take the same example *name* variable above and see how we can work with it using the Python's inbuilt <kbd>json</kbd> module. 

```python
import json
name = {
        "first_name" : "John", 
        "last_name" : "Doe"
       }

name_json_s = json.dumps(name)
print (name_json_s)
print ( type ( name_json_s) )
```

All the <kbd>dumps</kbd> method does is convert the Python dictionary object to a string. Let's work the other way round now.

Now, imagine you are given a string like so 
<pre>{"first_name": "John", "last_name": "Doe"}
</pre>

What if you want to convert it into a Python dictionary data type ? Use the <kbd>loads</kbd> or the 

```python
name_text = '{"first_name": "John", "last_name": "Doe"}'
name_json = json.loads(name_text)
print (name_json)
print ( type(name_json) )
```

Essentially, Python's <kbd>json</kbd> module is designed to convert json data in text format to Python's dictionary data type and vice-versa. 

<br>

<div class="text-center">
    
</div>

<br>

Also, we mentioned earlier that JSON data is essentially key value pairs. And the main property of a **key** is that they are unique. Just like the keys in a dictionary or a hash table. So, in case the JSON data base duplicate keys, python takes the latest key and ignores the previous occurrences. For example, say there are two first_names in the name json - 
- John
- James

```python
name_text = '''{
                "first_name": "John", 
                "first_name": "James",
                "last_name": "Doe"}
            '''
name_json = json.loads(name_text)
print (name_json)
```

Python loads up the JSON with the last key. JSON standard itself does not enforce unique keys. It is upto the interpreter - in this case Python - to process the data any way they like. Same is the behaviour of most languages including JavaScript.

<div class="alert alert-secondary" id="Format-JSON">    
    <span class="topic"> &#10070; Formatting JSON data in Python</span>   
</div>

<br>

JSON data can get pretty large. Here is a simple JSON document that represents the weather data of Mountain View on a certain day.

<pre>
{"coord": {"lon": -122.08, "lat": 37.39}, "weather": [{"id": 800, "main": "Clear", "description": "clear sky", "icon": "01d"}], "base": "stations", "main": {"temp": 282.55, "feels_like": 281.86, "temp_min": 280.37, "temp_max": 284.26, "pressure": 1023, "humidity": 100}, "visibility": 16093, "wind": {"speed": 1.5, "deg": 350}, "clouds": {"all": 1}, "dt": 1560350645, "sys": {"type": 1, "id": 5122, "message": 0.0139, "country": "US", "sunrise": 1560343627, "sunset": 1560396563}, "timezone": -25200, "id": 420006353, "name": "Mountain View", "cod": 200}
</pre>

It is just a long string that looks like a deeply nested dictionary. Let's load it into Python and see how the dictionary looks like. 

```python
weather = '''{"coord": {"lon": -122.08, "lat": 37.39}, 
              "weather": [{"id": 800, "main": "Clear", "description": "clear sky", "icon": "01d"}], 
              "base": "stations", 
              "main": {"temp": 282.55, "feels_like": 281.86, "temp_min": 280.37, "temp_max": 284.26, "pressure": 1023, "humidity": 100}, 
              "visibility": 16093, 
              "wind": {"speed": 1.5, "deg": 350}, 
              "timezone": -25200, "id": 420006353, "name": "Mountain View", "cod": 200}'''
weather_json = json.loads(weather)
print (weather_json)
```

Not much different, right ? That is because, Python prints out the JSON object (dictionary) into a string without any formatting. Let's try to convert the dictionary back to a JSON and see if it makes any difference. 

```python
print (json.dumps(weather_json))
```

No difference. This is because that is how the raw JSON data is transmitted across the internet. However, to make the JSON data more readable, Python can indent this data for us.

```python
print (json.dumps(weather_json, indent=4))
```

This looks much better, right ? You can very well convert this indented JSON string back to Python dictionary just as easily.

```python
weather_json = '''
{
    "coord": {
        "lon": -122.08,
        "lat": 37.39
    },
    "weather": [
        {
            "id": 800,
            "main": "Clear",
            "description": "clear sky",
            "icon": "01d"
        }
    ],
    "base": "stations",
    "main": {
        "temp": 282.55,
        "feels_like": 281.86,
        "temp_min": 280.37,
        "temp_max": 284.26,
        "pressure": 1023,
        "humidity": 100
    },
    "visibility": 16093,
    "wind": {
        "speed": 1.5,
        "deg": 350
    },
    "timezone": -25200,
    "id": 420006353,
    "name": "Mountain View",
    "cod": 200
}'''

weather_dict = json.loads(weather_json)
print ( weather_dict )
```

Now we know how to format large JSON documents in a human readable format.

<div class="alert alert-secondary" id="Nested-JSON">    
    <span class="topic"> &#10070; Nested JSON</span>   
</div>

<br>

JSON data can be nested. For example, the weather data above is an example of a nested JSON document. 

```python
{
    "coord": {
        "lon": -122.08,
        "lat": 37.39
    },
    
    "weather": [
        {
            "id": 800,
        },
        {
            "state" : "California"
        }
    ],
    "base": "stations",
    "main": {
        "temp": [282.55, 282.65],
    },
    "visibility": 16093,
    "wind": {
        "speed": 1.5,
        "deg": 350
    },
    "name": "Mountain View",
}
```

They key is always a string. However, the values could be any of the following.

- nested json object
- number or string
- an array of any of the above
