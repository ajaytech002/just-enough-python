```python
import tkinter as tk
# from tkinter import ttk

# main window
window = tk.Tk()
window.title("TODO App")

todo_text = tk.Entry(window)
todo_text.grid(row=0, column = 0)

add_button = tk.Button(window, text="Add")

add_button.grid(row=0, column = 1)


## Add a horizontal line
separator = tk.ttk.Separator(window)

## The width of the line is 1 px and length

################## START CHANGE #####################

separator.grid(row=1,columnspan=2, sticky="ew")

################# END CHANGE ########################
window.mainloop()
```

```python
import tkinter as tk
# from tkinter import ttk

##################### START CHANGE ####################
# List to hold the todo's that the user has added

todo_list = [] # initialize it to empty list.
#################### END CHANGE #######################

# main window
window = tk.Tk()
window.title("TODO App")


todo_string = tk.StringVar()

todo_text = tk.Entry(window, textvariable = todo_string)

todo_text.grid(row=0, column = 0)


def addClicked () :
    print ("todo = ",todo_string.get())
    ##################### START CHANGE ####################
    todo_list.append(todo_string.get())

    # just print the list every time to verify that the list is actually
    # getting saved
    print (" ".join(todo_list))
    #################### END CHANGE #######################
    
add_button = tk.Button(window, text="Add", command=addClicked)

add_button.grid(row=0, column = 1)


## Add a horizontal line
separator = tk.ttk.Separator(window)

## The width of the line is 1 px and length

separator.grid(row=1,columnspan=2, sticky="ew", pady=10)

window.mainloop()
```

We do see the scrollbar, but it is not doing anything when we do the scrolling. That is because we have to configure the list to use the scrollbar when we do the scrolling. It is done using the following code change. 

```python
import tkinter as tk
# from tkinter import ttk

todo_list = ["Lunch at 12:00 today", "Meeting preparation",
             "Third Item", "Fourth Item", "Fifth Item",
             "Sixth Item", "Seventh Item", "Eighth Item",
             "Ninth Item", "Tenth Item", "Eleventh Item",
             "Twelveth Item", "Thirteenth Item"] 

# main window
window = tk.Tk()
window.title("TODO App")


todo_string = tk.StringVar()

todo_text = tk.Entry(window, textvariable = todo_string)

todo_text.grid(row=0, column = 0)


def addClicked () :
    todo_list.append(todo_string.get())

add_button = tk.Button(window, text="Add", command=addClicked)

add_button.grid(row=0, column = 1)


## Add a horizontal line
separator = tk.ttk.Separator(window)

## The width of the line is 1 px and length

separator.grid(row=1,columnspan=2, sticky="ew", pady=10)

# Add the list of todo items from the list.
todo_list_display = tk.Listbox(window)

for todo in todo_list :
    todo_list_display.insert(END, todo)
    
todo_list_display.grid(row=2,column=0)


# Add a scrollbar
todo_list_scrollbar = tk.Scrollbar(window)
todo_list_scrollbar.grid(row=2,column=1)

#################### START CHANGE ###########################

# Attach the list to the scrollbar
todo_list_display.config(yscrollcommand=todo_list_scrollbar)
todo_list_scrollbar.config(command=todo_list_display.yview)
#################### END CHANGE #############################

window.mainloop()
```

In this section, we will add the buttons to make both of the following lists visible separately.
 - <kbd>open</kbd>
 - <kbd>completed</kbd>
 
To do this, we will create two buttons at the bottom that will show either of these lists. As soon as either of the buttons are clicked, the existing sets of todo widgets should be deleted and recreated depending on which button has been clicked. 

```python
import tkinter as tk
from tkinter import ttk

todo_list = ["Lunch at 12:00 today", "Meeting preparation",
             "Third Item", "Fourth Item", "Fifth Item",
             "Sixth Item", "Seventh Item", "Eighth Item",
             "Ninth Item", "Tenth Item", "Eleventh Item",
             "Twelveth Item", "Thirteenth Item", "Fourteenth Item"]
todo_list_complete = []

# This list will hold the GUI widgets that hold all the
# todos on the screen.
todo_list_display = []
open_button = None
completed_button = None
TODO_LIST_STATE = "OPEN"  # By default, we start the program with OPEN items

# main window
window = tk.Tk()
window.title("TODO App")


todo_string = tk.StringVar()

todo_text = tk.Entry(window, textvariable=todo_string)

todo_text.grid(row=0, column=0, padx=10, pady=10)


def addClicked():
    global TODO_LIST_STATE
    todo = todo_string.get()
    if todo == "":
        return
    todo_list.append(todo)
    todo_string.set("")  # clear the text box

    if TODO_LIST_STATE == "COMPLETED":
        TODO_LIST_STATE = "OPEN"

    render_list_items(todo_list)


add_button = tk.Button(window, text="Add", command=addClicked)

add_button.grid(row=0, column=1, padx=10, pady=10)


# Add a horizontal line
separator = ttk.Separator(window)

# The width of the line is 1 px and length

separator.grid(row=1, columnspan=2, sticky="ew", pady=10)


def mark_complete_clicked(index):

    # add to todo_list_complete
    todo_list_complete.append(todo_list[index])

    # remove the clicked element from todo_list
    del todo_list[index]

    # re-render the list items
    render_list_items(todo_list)


def mark_delete_clicked(index):
    del todo_list_complete[index]
    render_list_items(todo_list_complete)


def populate_open_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "OPEN"
    render_list_items(todo_list)


def populate_completed_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "COMPLETED"
    render_list_items(todo_list_complete)

# Add the list of todo items from the list.


def render_list_items(todo_list):

    # if there are existing todo list items (widgets), destroy them
    for index, entry in enumerate(todo_list_display):
        # entry[0].set('Hello') # cannot destroy string vars.
        entry[1].destroy()
        entry[2].destroy()

    todo_list_display.clear()

    # by default variables are local in Python.
    global open_button
    global completed_button

    # destroy the buttons at the bottom
    if open_button != None:
        open_button.destroy()
    if completed_button != None:
        completed_button.destroy()

    row = 2

    for index, todo in enumerate(todo_list):
        todo_list_entry = []

        todo_list_entry.append(tk.StringVar())

        todo_list_entry.append(
            tk.Entry(window, textvariable=todo_list_entry[0]))

        todo_list_entry[0].set(todo)
        todo_list_entry[1].config(state="disabled")
        row = row + 1
        todo_list_entry[1].grid(row=row, column=0, padx=10, pady=3)

        text = ""
        if TODO_LIST_STATE == "OPEN":
            text = "✔"
            todo_list_entry.append(tk.Button(window, text=text,
                                             command=lambda index=index: mark_complete_clicked(index)))
        elif TODO_LIST_STATE == "COMPLETED":
            text = "❌"
            todo_list_entry.append(tk.Button(window, text=text,
                                             command=lambda index=index: mark_delete_clicked(index)))

        todo_list_entry[2].grid(row=row, column=1, padx=10, pady=3)

        todo_list_display.append(todo_list_entry)

    # Since we don't know the row numbers dynamically, add the buttons here only

    open_button = tk.Button(window, text="Open", command=populate_open_items)
    open_button.grid(row=row+1, columnspan=2,
                     sticky="W", padx=10, pady=(10, 10))

    completed_button = tk.Button(
        window, text="Completed", command=populate_completed_items)
    completed_button.grid(row=row+1,  columnspan=2, pady=(10, 10))


render_list_items(todo_list)


window.mainloop()
```

What about scrolling ? If the list of items exceed the window height limit, you are not going to see the overflowing items. Naturally, we need to add a scroll bar. But which widget should we associate it to ? We have a whole bunch of items that we dynamically generate. 

<kbd>Scrollbar</kbd> widget is easy to create. However, there is a hard limitation of the <code>Scrollbar</code> widget that will not allow us to use it in our situation. Let's solve that issue using the concept of <kbd>Frame</kbd> and <kbd>Canvas</kbd>. 

```python
import tkinter as tk
from tkinter import ttk

todo_list = ["Lunch at 12:00 today", "Meeting preparation",
             "Third Item", "Fourth Item", "Fifth Item",
             "Sixth Item", "Seventh Item", "Eighth Item",
             "Ninth Item", "Tenth Item", "Eleventh Item",
             "Twelveth Item", "Thirteenth Item", "Fourteenth Item"]
todo_list_complete = []

# This list will hold the GUI widgets that hold all the
# todos on the screen.
todo_list_display = []
open_button = None
completed_button = None
TODO_LIST_STATE = "OPEN"  # By default, we start the program with OPEN items

# main window
window = tk.Tk()
window.title("TODO App")

# Scrollbars do not work with a group of widgets. They
# are only supported out of the box on certain items
# like Listbox etc. So, we have to have a "canvas" that can
# be made scrollable.

canvas = tk.Canvas(window, bg="#333", 
                   width=200, 
                   height=300)
canvas.grid(row=2, column=0)

# Frame to hold todo list items (open)
frame = tk.Frame(canvas, bg="#555")
frame.grid(row=0, column=0, pady=10)


frame.bind("<Configure>", 
    lambda e: canvas.configure(
    scrollregion=canvas.bbox("all")))

# Scrollbar
scrollbar = tk.Scrollbar(window, 
                         orient="vertical", 
                         command=canvas.yview)
scrollbar.grid(column=1, 
               row=2, sticky="ns")

canvas.configure(yscrollcommand=scrollbar)
canvas.create_window(10, 10, 
                     window=frame, 
                     anchor=tk.NW)

todo_string = tk.StringVar()

todo_text = tk.Entry(window, textvariable=todo_string)

todo_text.grid(row=0, column=0, padx=10, pady=10)


def addClicked():
    global TODO_LIST_STATE
    todo = todo_string.get()
    if todo == "":
        return
    todo_list.append(todo)
    todo_string.set("")  # clear the text box

    if TODO_LIST_STATE == "COMPLETED":
        TODO_LIST_STATE = "OPEN"

    render_list_items(todo_list)


add_button = tk.Button(window, text="Add", command=addClicked)

add_button.grid(row=0, column=1, padx=10, pady=10)


# Add a horizontal line
separator = ttk.Separator(window)

# The width of the line is 1 px and length

separator.grid(row=1, columnspan=2, sticky="ew", pady=10)


def mark_complete_clicked(index):

    # add to todo_list_complete
    todo_list_complete.append(todo_list[index])

    # remove the clicked element from todo_list
    del todo_list[index]

    # re-render the list items
    render_list_items(todo_list)


def mark_delete_clicked(index):
    del todo_list_complete[index]
    render_list_items(todo_list_complete)


def populate_open_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "OPEN"
    render_list_items(todo_list)


def populate_completed_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "COMPLETED"
    render_list_items(todo_list_complete)

# Add the list of todo items from the list.


def render_list_items(todo_list):

    for index, entry in enumerate(todo_list_display):
        entry[1].destroy()
        entry[2].destroy()

    todo_list_display.clear()

    # by default variables are local in Python.
    global open_button
    global completed_button

    row = 0

    for index, todo in enumerate(todo_list):
        todo_list_entry = []

        todo_list_entry.append(tk.StringVar())

        todo_list_entry.append(
            tk.Entry(frame, textvariable=todo_list_entry[0]))

        todo_list_entry[0].set(todo)
        todo_list_entry[1].config(state="disabled")
        row = row + 1
        todo_list_entry[1].grid(row=row, column=0, padx=10, pady=3)

        text = ""
        if TODO_LIST_STATE == "OPEN":
            text = "✔"
            todo_list_entry.append(tk.Button(frame, text=text,
                                             command=lambda index=index: mark_complete_clicked(index)))
        elif TODO_LIST_STATE == "COMPLETED":
            text = "❌"
            todo_list_entry.append(tk.Button(frame, text=text,
                                             command=lambda index=index: mark_delete_clicked(index)))

        todo_list_entry[2].grid(row=row, column=1, padx=10, pady=3)

        todo_list_display.append(todo_list_entry)


render_list_items(todo_list)

open_button = tk.Button(window, text="Open", command=populate_open_items)
open_button.grid(row=3, columnspan=2,
                 sticky="W", padx=10, pady=(10, 10))

completed_button = tk.Button(
    window, text="Completed", command=populate_completed_items)
completed_button.grid(row=3,  columnspan=2, pady=(10, 10))

window.mainloop()
```

Just to make things easy to understand, we have created the <kbd>Canvas</kbd> and <kbd>Frame</kbd> as follows. 

- <kbd>Canvas</kbd> - painted black 
- <kbd>Frame</kbd> - painted grey

This should give you a good idea on how they are created. As to the why, 

- A bunch of widgets are NOT scrollable. To circumvent this problem, we use a Canvas widget. 
- However, we can only associate one widget with a Canvas. To go around this problem, we use a Frame widget that associates all the todo list widgets to the Frame widget and the Frame widget will be associated with the Canvas object.

Also, see how the frame is setup as a child of the Canvas widget. That is why the row=0 for the Frame Widget. So, it is no longer following the main window's grid.

How do we retrieve data items though ? With the current data structure - <kbd>todo_list</kbd> we cannot uniquely identify the todo text because, we we are not storing the unique id when fetching the todo items durig the initial program run. In order to do that, we need a better way of storing data in the list. 

Instead of storing the todo items in a flat list, let's store them in a list of lists. That way we can store all of the data associated with a todo item - like its unique database id and if necessary the status as well. Here is a visual of the data structure change.

<br>

</div>

<br>

To accomodate for this data structure change, the following changes need to be done in the program. 

```python
import tkinter as tk
import tkinter.font as font
from tkinter import ttk
import sqlite3

# Constants
DATABASE = "todo.db"
TABLE_TODO_ITEMS = "todo_items"

# DATABASE operations #############################3


def todo_items_fetch(status):
    try:
        data = []
        conn = sqlite3.connect(DATABASE)
        cursor = conn.execute(
            '''SELECT id, description, status FROM todo_items
               WHERE status = ?''', [status])
        for row in cursor:
            data.append(row)
        return data
    except Exception as e:
        print("could not get data from table")
        print("error =>", e)
    finally:
        conn.close()


def add_todo_item(todo_item):
    try:
        conn = sqlite3.connect(DATABASE)
        conn.execute('''INSERT INTO todo_items(description,status) 
                        VALUES (?,"OPEN")''', [todo_item])
        conn.commit()
    except Exception as e:
        print("error =>", e)
    finally:
        conn.close()


def set_status_todo_item(id, status):
    try:
        conn = sqlite3.connect(DATABASE)
        conn.execute(
            '''UPDATE todo_items SET status = ? WHERE id = ?''', [status, id])
        conn.commit()
    except Exception as e:
        print("error=> ", e)
    finally:
        conn.close()

#############################################################


# This list will hold the GUI widgets that hold all the
# todos on the screen.
todo_list_display = []
open_button = None
completed_button = None
TODO_LIST_STATE = "OPEN"  # By default, we start the program with OPEN items

# main window
window = tk.Tk()
window.title("TODO App")

canvas = tk.Canvas(window, width=200, height=300)
canvas.grid(row=2, column=0, sticky="ns")

# Frame to hold todo list items (open)
frame = tk.Frame(canvas)
frame.grid(row=0, column=0, pady=10)
frame.bind("<Configure>", lambda e: canvas.configure(
    scrollregion=canvas.bbox("all")))

# Scrollbar
scrollbar = tk.Scrollbar(
    window, orient="vertical", command=canvas.yview)
scrollbar.grid(column=1, row=2, sticky="ns")

canvas.configure(yscrollcommand=scrollbar)
canvas.create_window(10, 10, window=frame, anchor=tk.NW)

todo_string = tk.StringVar()

todo_text = tk.Entry(window, textvariable=todo_string)

todo_text.grid(row=0, column=0, padx=10, pady=10)


def addClicked():
    global TODO_LIST_STATE
    todo = todo_string.get()
    if todo == "":
        return
    todo_string.set("")  # clear the text box

    if TODO_LIST_STATE == "COMPLETED":
        TODO_LIST_STATE = "OPEN"

    add_todo_item(todo)  # add the todo item to the db
    todo_list = todo_items_fetch("OPEN")  # fetch all the OPEN items again
    render_list_items(todo_list)  # render all the OPEN items again


add_button = tk.Button(window, text="Add", command=addClicked)

add_button.grid(row=0, column=1, padx=10, pady=10)


# Add a horizontal line
separator = ttk.Separator(window)

# The width of the line is 1 px and length

separator.grid(row=1, columnspan=2, sticky="ew", pady=10)


def mark_complete_clicked(id):
    # mark the item as COMPLETED
    set_status_todo_item(id, "COMPLETED")
    # re-render the OPEN list items
    populate_open_items()


def mark_delete_clicked(id):
    set_status_todo_item(id, "DELETED")
    # re-render the COMPLETED list items
    populate_completed_items()


def populate_open_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "OPEN"
    todo_list = todo_items_fetch("OPEN")
    render_list_items(todo_list)


def populate_completed_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "COMPLETED"
    todo_list = todo_items_fetch("COMPLETED")
    render_list_items(todo_list)


def render_list_items(todo_list):
    global todo_list_display

    for index, entry in enumerate(todo_list_display):
        entry[2].destroy()
        entry[3].destroy()

    # contains the widgets and other GUI elements that
    # are needed for displaying the todo list items.
    todo_list_display.clear()

    # by default variables are local in Python.
    global open_button
    global completed_button

    row = 0

    for index, todo in enumerate(todo_list):
        todo_list_entry = []
        todo_list_entry.append(todo[0])  # unique id - index 0
        todo_list_entry.append(tk.StringVar())  # StrinvVar - index 1

        if TODO_LIST_STATE == "OPEN":
            font_type = font.Font(family="Helvetica", size=10)
        elif TODO_LIST_STATE == "COMPLETED":
            font_type = font.Font(family="Helvetica", size=10, overstrike=1)

        todo_list_entry.append(
            tk.Entry(frame, textvariable=todo_list_entry[1], font=font_type))  # textbox - index 2

        # set the StringVar to the actual todo text
        todo_list_entry[1].set(todo[1])
        todo_list_entry[2].config(state="disabled")
        row = row + 1
        todo_list_entry[2].grid(row=row, column=0, padx=10, pady=3)

        text = ""
        if TODO_LIST_STATE == "OPEN":
            text = "✔"
            todo_list_entry.append(tk.Button(frame, text=text,
                                             command=lambda index=todo[0]: mark_complete_clicked(index)))
        elif TODO_LIST_STATE == "COMPLETED":
            text = "❌"
            todo_list_entry.append(tk.Button(frame, text=text,
                                             command=lambda index=todo[0]: mark_delete_clicked(index)))

        todo_list_entry[3].grid(row=row, column=1, padx=10, pady=3)

        todo_list_display.append(todo_list_entry)


### INITIAL RENDER###
todo_list = todo_items_fetch("OPEN")
render_list_items(todo_list)

# OPEN BUTTON AT THE BOTTOM
open_button = tk.Button(window, text="Open", command=populate_open_items)
open_button.grid(row=3, columnspan=2,
                 sticky="W", padx=10, pady=(10, 10))

# COMPLETED BUTTON AT THE BOTTOM
completed_button = tk.Button(
    window, text="Completed", command=populate_completed_items)
completed_button.grid(row=3,  columnspan=2, pady=(10, 10))

window.mainloop()
```
